```python
import numpy as np
import pandas as pd
import geopandas as gpd
from shapely import wkt
```

```python
data = pd.read_csv("Affordable_Housing_Developments.csv")
```

```python
data.head()
```

```python
data['geometry'] = data['New Georeferenced Column'].astype(str)
data['geometry'] = data['geometry'].str.strip()
data['geometry'] = data['geometry'].apply(wkt.loads)
gdf = gpd.GeoDataFrame(data, crs='epsg:4326')
gdf
```

```python
gdf_2 = gdf.drop(columns = ['New Georeferenced Column'])
gdf_2
```

```python
gdf_2.to_file("affordable_housing.shp")
```
